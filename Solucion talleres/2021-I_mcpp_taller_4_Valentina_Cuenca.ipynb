```python
import math 
def SpheraArea(r): 
    pi= math.pi
    area= 4 * pi * (r**2)
    return area
SpheraArea(6)
```

```python
def SpheraVolume(r): 
    pi= math.pi
    Volumen= (4 * pi * (r**3))/3
    return Volumen
SpheraVolume(6)
```

##### 4.

```python
def sumN(n): 
    suma=0
    for i in range(1,n+1):
        suma=suma+i
    return suma 
def sumNCubes(n):
    suma=0
    for i in range(1,n+1):
        suma = suma + i**2
    return suma 
```

```python
n=int(input("Escriba un numero: "))
sumN(n)
```

```python
sumNCubes(n)
```

```python
def SquareEach(nums):
    resultado=[]
    for i in nums:
        resultado.append(i**2)
    return resultado
SquareEach([1,2,3,4])
```

```python
def sumList(nums): 
    resultado=0
    for i in nums: 
        resultado = i + resultado
    return resultado 
sumList([1,2,3,4])
```
