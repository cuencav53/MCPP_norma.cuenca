### 1.
Calcule el número de crímenes en cada Community Area en 2015. Haga un gráfico de barras que lo ilustre.

```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np

crimes = pd.read_csv('Crimes_2015.csv')
crimes
```

```python
crimes_by_Comunity = crimes.groupby("Community Area")
crimes_by_Comunity['ID'].agg('count')
```

```python
plt.rcParams["figure.figsize"] = [18.0, 8.0]
crimes_by_Comunity['ID'].agg('count').plot(kind='bar');
```

### 2.
Ordene las Community Areas de acuerdo con el número de crímenes. ¿Qué Community Area (por nombre, idealmente) presenta el mayor número de crímenes? ¿El menor?

```python
cimesC=crimes_by_Comunity['ID'].agg('count').to_frame()
cimesC.sort_values("ID")
```

```python
cimesC.sort_values("ID", ascending=False)
```

El que presenta el mayor número de crimenes es el 25 y el menos es el 9.

### 4.

Descargue la base de datos de información socioeconómica (https://data.cityofchicago.org/Health-Human-Services/Census-Data-Selected-socioeconomic-indicators-in-C/kn9c-c2s2).

Cree una tabla que agregue el número de crímenes por Community Area. Una esa tabla con la de datos socioeconómicos y cree un "scatter plot" de número de crímenes vs ingreso per cápita. Explique la relación en palabras.

```python
soec = pd.read_csv("Socioeconomico.csv")
soec
```

```python
crimes_by_community = crimes.groupby('Community Area')
community_crime_count = crimes_by_community['ID'].agg('count')
community_crime_count=community_crime_count.to_frame()
community_crime_count = community_crime_count.reset_index()
community_crime_count
```

```python
soec_crimenes=pd.merge(soec, community_crime_count, how='left', left_index=True, right_index=True)
soec_crimenes=soec_crimenes.rename(columns={'ID':'Crimenes'})
```

```python
soec_crimenes
```

```python
crimenes_pib=soec_crimenes[["Community Area",'PER CAPITA INCOME ',"Crimenes"]]
```

```python
crimenes_pib.plot.scatter (x = 'PER CAPITA INCOME ', y = "Crimenes");
```

Podemos darnos cuenta que para las areas que tienen un alto PIB per-capita son pocas, pero tienen un crimenes considerables, y los que tienen un PIB per-capita mas bajo, no tienen crimenes tan altos, por lo que se puede deducir una relacion positiva no muy fuerte.
