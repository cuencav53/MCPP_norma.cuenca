4. Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?

```python
from nltk.corpus import state_union
```

```python
palabras_clave = nltk.ConditionalFreqDist((objetivo, documentos[:4])
          for documentos in state_union.fileids()
          for w in state_union.words(documentos)
          for objetivo in ['men', 'women', 'people']
          if w.lower().startswith(objetivo))
palabras_clave.plot();
```

Podemos ver que la palabra people cuando mas se utilizó fue entre 1992 y 1997, mientras que la palabra men se utiliza menos, en comparqción a people,  a lo largo de todos los años, pero se utiliza más si la comparamos con la palabra women, en el 2005 y 2002 women supera a men, y en el 2006 las dos palabras se utilizan en la misma cantidad. Tambien es interesante entre 1965 y 1968 como men y people aumentaron al mismo tiempo.
