1. Realiza una función separar(lista) que tome una lista de números enteros y devuelva dos listas ordenadas. La primera con los números pares y la segunda con los números impares.

```python
numeros = [-12, 84, 13, 20, -33, 101, 9]
def separar(lista):
    pares=[]
    impares=[]
    for i in lista:
        if i%2==0:
            pares.append(i)
        else:
            impares.append(i)
    pares.sort()
    impares.sort()
    return pares, impares
a = separar(numeros)
list_Pares=a[0]
list_Impares=a[1]
```

```python
list_Pares, list_Impares = separar(numeros)
```

```python
print(list_Pares)
print(list_Impares)
```

2. Crea una función que calcule el factorial de un número pasado por parámetro

```python
def factorial(num):
    resultado = num
    if num==0:
        return 1
    for i in range(num-1, 1, -1):
        resultado= resultado*i
    return resultado
```

```python
print(factorial(5))
```

3. Crea una función que dados dos números mostrará todos los números que hay entre ellos.

```python
if num1>num2:
    aux = num1
    num1 = num2
    num2 = aux
for i in range(num1+1, num2):
    print(i, " ", end="")
```

```python

def numerosEntre(num1, num2):
    if num1>num2:
        aux = num1
        num1 = num2
        num2 = aux
    for i in range(num1+1, num2):
        print(i, " ", end="")

numerosEntre(15, 10)
```
