```python
def separa(n): 
    par=[]
    impar=[]
    for i in range(len(n)): 
        if n[i]%2==0: 
            par.append(n[i])
        else: 
            impar.append(n[i])
     
    par.sort()
    impar.sort()
    return par, impar

n=[1,2,3,6,5,3,9]
separa(n)
```

```python
def factorial(m): 
    x=1
    while m!=1 and m!=0: 
        x=m*x
        m=m-1
    print(x)
factorial(1)
```

```python
def entre(n,m): 
    x=min(n,m)
    y=max(n,m)
    z=[]
    for i in range(x,y+1):
        z.append(i)
    print(z)
        
entre(2,6)
```

```python
import random
def juego():
    jugar="si"
    while jugar=="si":
        resultado=[]
        #Computador
        computador=random.randint(0,2)
        if computador==0: 
            computadorn= "Piedra"
        elif computador==1: 
            computadorn="Papel"
        else: 
            computadorn="Tijera"
        #Jugador
        jugadorn=input("¿ Piedra, papel o tijera? (p/a/t): ")

        #Resultado
        resultado.append(computadorn)
        resultado.append(jugadorn)
        print("El computador escogio: ",computadorn)
        if resultado== ["Piedra","t"] or resultado==["Papel","p"] or resultado==["Tijera","a"]:
            print("¡Has perdido!")
        elif resultado== ["Piedra","a"] or resultado==["Papel","t"] or resultado==["Tijera", "p"]:
            print("¡Has ganado!")
        else:
            print("Empate")
        jugar=input("¿Quieres volver a jugar? (si/no): ")
        if jugar=="no":
            print("Fin del juego, adios.")
        
juego()
    
```

```python
import random

def  buscar_bolita(inicio):
    uno= False 
    dos= False 
    tres= False 
    if inicio== "1":
        uno= True 
    elif inicio=="2":
        dos=True
    else: 
        tres=True 
    for i in range(3):
        final=random.randint(1,4)
        print(final)
        if final==1:
            sustituto=uno
            uno=tres
            tres=sustituto
        elif final ==2 : 
            sustituto=uno
            uno=dos
            dos=sustituto
        else : 
            sustituto=dos
            dos=tres
            tres=sustituto
    if uno== True: 
        resultado="1"
    elif dos== True: 
        resultado= "2"
    else: 
        resultado="3"
    return resultado 

def es_ganador(resultado,final):
    if resultado==final: 
        print("Ganó")
    else:
        print("Perdió")

        
inicio=input("Escoja el vaso donde va a iniciar la bolita (1/2/3): ")
final=input("Escoja el vaso donde va a terminar la bolita (1/2/3): ")

resultado=buscar_bolita(inicio)
es_ganador(resultado,final)
        
    
        
```

```python
def creart(n):
    n=str(n)
    nombre="tabla_"+n+".txt"
    a=open(nombre,"w")
    for i in range(11):
        n=int(n)
        r=str( n*i)
        i=str(i)
        n=str(n)
        a.write(i)
        a.write("*")
        a.write(n)
        a.write("=")
        a.write(r )
        a.write("\n")
        
    a.close()

    
creart(8)
def verificar(n):
    n=str(n)
    nombre="tabla_"+n+".txt"
    try:
        a=open(nombre,"r")
    except FileNotFoundError:
        print('No existe el fichero con la tabla del', n)
    else:
        print(a.read())
verificar(8)
```
