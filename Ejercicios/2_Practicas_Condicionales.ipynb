1. Escriba un programa que pida dos números enteros. El programa pedirá de nuevo el segundo número mientras no sea mayor o igual que el primero. El programa terminará escribiendo los dos números.


```python
primero= input( "Digame un número: ")
segundo= input("Digame otro número mayor que el anterior: " )
if segundo > primero: 
    print( "sus numeros fueron", primero, "y", segundo)
while segundo < primero: 
    print( segundo, "no es mayor que", primero)
    segundo= input( "Intentelo de nuevo: ")
print( "sus numeros fueron", primero, "y", segundo)
    
```

2. Escriba un programa que pida la anchura de un triángulo y lo dibuje con caracteres producto (*)


```python
n= input("Digame la anchura: ")
n=int(n)
for i in range(n+1):
    print("*"*i)
for j in range(n-1,0,-1):
    print("*"*j)
```
