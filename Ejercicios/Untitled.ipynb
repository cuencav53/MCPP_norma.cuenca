```python
def separa(n): 
    par=[]
    impar=[]
    for i in range(len(n)): 
        if n[i]%2==0: 
            par.append(n[i])
        else: 
            impar.append(n[i])
     
    par.sort()
    impar.sort()
    return par, impar

n=[1,2,3,6,5,3,9]
separa(n)
```

```python
def factorial(m): 
    x=1
    while m!=1 and m!=0: 
        x=m*x
        m=m-1
    print(x)
factorial(1)
```

```python
def entre(n,m): 
    x=min(n,m)
    y=max(n,m)
    z=[]
    for i in range(x,y+1):
        z.append(i)
    print(z)
        
entre(2,6)
```

```python
import random
def juego():
    jugar="si"
    while jugar=="si":
        resultado=[]
        #Computador
        computador=random.randint(0,2)
        if computador==0: 
            computadorn= "Piedra"
        elif computador==1: 
            computadorn="Papel"
        else: 
            computadorn="Tijera"
        #Jugador
        jugadorn=input("¿ Piedra, papel o tijera? (p/a/t): ")

        #Resultado
        resultado.append(computadorn)
        resultado.append(jugadorn)
        print("El computador escogio: ",computadorn)
        if resultado== ["Piedra","t"] or resultado==["Papel","p"] or resultado==["Tijera","a"]:
            print("¡Has perdido!")
        elif resultado== ["Piedra","a"] or resultado==["Papel","t"] or resultado==["Tijera", "p"]:
            print("¡Has ganado!")
        else:
            print("Empate")
        jugar=input("¿Quieres volver a jugar? (si/no): ")
        if jugar=="no":
            print("Fin del juego, adios.")
        
juego()
    
```
