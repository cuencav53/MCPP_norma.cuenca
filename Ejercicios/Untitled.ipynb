```python
def separa(n): 
    par=[]
    impar=[]
    for i in range(len(n)): 
        if n[i]%2==0: 
            par.append(n[i])
        else: 
            impar.append(n[i])
     
    par.sort()
    impar.sort()
    return par, impar

n=[1,2,3,6,5,3,9]
separa(n)
```
