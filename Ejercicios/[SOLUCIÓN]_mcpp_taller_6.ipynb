### 4.1 Plot a simple graph of a sinus function in the range 0 to 3 with a step size of 0.01

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 3.0, 0.01)
h = np.sin(2*np.pi*t)
plt.scatter(t,h)
plt.show()
```

### 4.2 Make the line red. Add diamond-shaped markers with size of 5

```python
plt.scatter(t,h, s=5, color='r', marker="D")
```

### 4.3 Add a legend and a grid to the plot

```python
plt.scatter(t,h, s=5, color='r', label='sin(x)', marker="D")

plt.grid(True)
plt.legend();
```

### 5.1 Apply different line styles to a plot. Change line color and thickness as well as the size and the kind of the marker. Experiment with different styles

### 6.1 Annotate a line at two places with text. Use green and red arrows and align it according to figure points and data

```python
%matplotlib inline
x = np.linspace(0, 10,100)
plt.plot(x, np.exp((-x)/10)*np.sin(np.pi*x), 'g',  label='f(x)')
plt.plot(x, x*np.exp((-x)/3), 'b', label='g(x)')
plt.title('Funciones')

plt.annotate('g(x)',  xy = (8, 1), color='r', arrowprops={'facecolor': 'b'}, verticalalignment='center')
plt.annotate('f(x)',  xy = (4, -0.65), color='g', arrowprops={'facecolor': 'b'}, horizontalalignment='left');
```

#### 1. Plot a graph with dates for one year with daily values at the x axis using the built-in module datetime

```python
import datetime 
from matplotlib.dates import DayLocator, HourLocator, DateFormatter, drange ,MonthLocator 
import matplotlib.dates as dt
from numpy import arange
import datetime
a = datetime.datetime(2019,1,1)
numdays = 365
dateList = []
for x in range (0, numdays):
    dateList.append(a - datetime.timedelta(days = x))
import random
y = [random.randint(-20,20) for x in range(365)]
plt.plot(dateList, y);
```

### 2. Format the dates in such a way that only the first day of the month is shown

```python
plt.gca().xaxis.set_major_formatter(DateFormatter(fmt="01-%b"))
plt.plot(dateList, y);
```

#### 3. Place a small plot in one bigger plot

```python
x = [i for i in range(101)]
y1 = [-i**2 for i in x]
y2 = [i**3 for i in x]

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.72, 0.72, 0.16, 0.16])

ax1.plot(x,y1)
ax2.plot(x,y2)

ax1.set_title('Cuadratica')
ax1.set_xlabel('Eje X')
ax1.set_ylabel('Eje Y')

ax2.set_title('Cuadratica')
ax2.set_xlabel('Eje X')
ax2.set_ylabel('Eje Y')
```
