2. Escribe un programa que lea una cadena y devuelva un diccionario con la cantidad de apariciones de cada palabra en la cadena. Por ejemplo, si recibe “Qué lindo día que hace hoy” debe devolver: ‘que’: 2, ‘lindo’: 1, ‘día’: 1, ‘hace’: 1, ‘hoy’: 1

```python
frase=input("Escriba una cadena: ")
contar={}
frase=frase.lower()
y=frase.split()

for i in y:
    if i not in contar: 
        contar[i]=0
    contar[i] = contar[i] + 1
for palabra,frecuencia in contar.items():
    print(palabra,"-->", frecuencia)
    
        
```
