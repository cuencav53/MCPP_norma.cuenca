```python
# Supongamos un programa para hacer triage en sala de emergencias
riesgo = "medio"
if riesgo == "alto":
    print("El riesgo del paciente es", riesgo)
    print("Asigne paciente a grupo A")
    print("Llame de inmediato a contacto de emergencia")
elif riesgo == "medio":
    print("El riesgo del paciente es", riesgo)
    print("Asigne paciente a grupo B")
else:
    print("El riesgo del paciente es", riesgo)
    print("Asigne paciente a grupo C")
print("Diagnóstico finalizado")
```

```python
es_primo = True
n = 7
for i in range(2, n):
    if n % i == 0:
        es_primo = False  # Si n es divisible por i, entonces n no es primo
print(n, "¿Es primo?", es_primo)
```

```python
es_primo = True
n = 8
for i in range(2, n):
    if n % i == 0:
        es_primo = False
print(n, "¿Es primo?", es_primo)
```

Cuando n = 8, el programa "funciona", pero es computacionalmente ineficiente.

```python
es_primo = True
n = 8
for i in range(2, n):
    if n % i == 0:
        es_primo = False
        break
print(n, "¿Es primo?", es_primo)
```

Usar "print statements" siempre es buena idea (en el proceso).

## <font face="consolas">while</font> loops

```python
i = 0
N = 10
while i < N:
    print(i)
    i = i + 1
```

```python
i = 0
N = 10
while i < N:
    i = i + 1
    print(i)
```
