En general, código de la forma:

```python
if key not in d:
    d[key] = some_default_value
d[key] = some_transformation(d[key])
```

puede ser escrito como:

```python
d[key] = some_transformation(d.get(key, some_default_value))
```

# "Functional programming"

### Funciones anónimas o "lambda functions"

```python
def square(x):
    return x**2
```

```python
nums
```

```python
for num in nums:
    print(square(num))
```

```python
f = lambda x: x**2
for num in nums:
    print(f(num))
```

### map( ), filter ( ), reduce ( )

```python
nums = [1,2,3,4,5]
```

```python
def incr(x):
    return x + 1
```

```python
map(incr, nums)
```

```python
list(map(incr, nums))
```

```python
list(map(lambda x: x+1, nums))
```

```python
list(map(lambda x: x**2, nums))
```

```python
def is_odd(x):
    return x % 2 == 1
```

```python
nums
```

```python
list(filter(is_odd, nums))
```

```python
list(filter(lambda x: x%2==1, nums))
```

```python
list(filter(lambda x: x%2==0, nums))
```

```python
g = lambda x,y,z: x*y*z
for num in nums:
    print(g(num,2,3))
```

```python
from functools import reduce
```

```python
def add(x, y):
    return x + y
```

```python
reduce(add, [1,2,3,4])
```

```python
reduce(lambda x,y: x*y, [1,2,3,4])
```

# "List comprehensions"

```python
list(map(lambda x: x**2, nums))
```

```python
nums2 = []
for x in nums:
    nums2.append(x**2)
```

```python
nums2
```

```python
y=[x**2 for x in nums]
```

```python
list(map(lambda x: x**2, filter(lambda x: x%2==1, nums)))
```

```python
nums3 = []
for x in nums:
    if x % 2 == 1:
        nums3.append(x**2)
```

```python
nums3
```

```python
[x**2 for x in nums if x%2==1]
```

```python
d = {x+1:0 for x in range(10)}
```
